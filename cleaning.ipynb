```python
import pandas as pd
```

```python
df = pd.read_csv("Superstore.csv")
```

```python
df = pd.read_csv("Superstore.csv", encoding="ISO-8859-1")
```

```python
df.head()
df.columns
```

```python
df = df.drop(columns=[
    "Row ID",
    "Order ID",
    "Customer ID",
    "Customer Name",
    "City",
    "Country",
    "Product ID",
    "Product Name"
])


```

```python
df.columns()
```

```python
df.head()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=False, errors="coerce")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=False, errors="coerce")
```

```python
df[["Order Date", "Ship Date"]].head(10)
```

```python
df["Delivery Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
df["Delivery Days"].head(10)
```

```python
df.head()
```

```python
df.columns
```

```python
df["Unit Cost"] = (df["Sales"] - df["Profit"]) / df["Quantity"]
```

```python
df[["Sales", "Profit", "Quantity", "Unit Cost"]].head()
```

```python

df["Order Month"] = df["Order Date"].dt.month


df["Order Year"] = df["Order Date"].dt.year
```

```python
df["Order Month Name"] = df["Order Date"].dt.month_name()
```

```python
df[["Order Date", "Order Month", "Order Year", "Order Month Name"]].head()
```

```python
df.to_csv("cleaned_superstore.csv", index=False)
```
